Implement a solution to the knapsack problem that uses backtracking. Your method receives three parameters:
- weights: list that contains the  weights of all the items in the house. That is, item *i* has a weight of *weights[i]*
- values: list that contains the values of all the items in the house. That is, item *i* has a value of *values[i]*
- limit: maximum capacity (weight) that the knapsack can hold.

Feel free to add helper methods if necessary. Just don't change the signature of the original method.

```
def solve_knapsack_problem(weights, values, limit):
  i = 0
  return recursive(weights, values, limit, 0, i)

def recursive(weights, values, remaining, currValue, i):
  if remaining < 0:
    return -1
  if i >= len(weights):
    return currValue

  a = recursive(weights, values, remaining-weights[i], currValue+values[i], i+1) #Added
  b = recursive(weights, values, remaining,currValue, i+1) #Not added

  if a > b:
    return a
  return b
```
