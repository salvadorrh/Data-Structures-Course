### [50 points] Problem 1

Write a method that, given a binary tree as input, finds the maximum sum you can compute by adding all the values from the root to one of the leaves. Note: Unlike a regular binary tree, for this problem, assume that nodes in the tree SHARE children (see example). Feel free to write helper methods. You have to use dynamic programming to get credit.

    Example:

           3
         /   \
        10    6
       /  \  /  \
      -5  -70   -4
     /  \ /  \ /  \
    -1   2   50   -8

    Output: 3 + 6 + (-4) + 50 = 55 


```
class Node:
  def __init__(self, left=None, right=None, item=None):
    self.left = left
    self.right = right
    self.item = item

def max_sum(root):
  if root is None:
    return 0
  seen = {}
  return helper(root, seen)

def helper(root, seen):
  if root is None:
    return 0
  left = 0
  right = 0 
  if root.left is not None:
    if root.left in seen:
      left = seen[root.left]
    else:
      left = helper(root.left, seen)   

  if root.right is not None:
    if root.right in seen:
      right = seen[root.right]
    else:
      right = helper(root.right, seen) 

  #print("Root: ", root.item,"Left: ", left, "Right: ", right)
  if left > right:
    seen[root] = root.item + left
    return seen[root]
  else:
    seen[root] = root.item + right
    return seen[root] 
```

### [50 points] Problem 2

Imagine an island that is in the shape of a bar graph. When it rains, certain areas of the island fill up with rainwater to form lakes. Any excess rainwater the island cannot hold in lakes will run off the island to the west or east and drain into the ocean.
Given a 1D array of positive integers representing 2D bar heights, write a function that computes the total volume (capacity) of water that could be held in all lakes on such an island given an array of the heights of the bars. Assume an elevation map where the width of each bar is 1. You have to use dynamic programming to get credit.

Example 1: Given [1,3,2,4,1,3,1,4,5,2,2,1,4,2,2], return 15 (3 bodies of water with volumes of 1,7,7 yields total volume of 15)

image.png

Example 2: Given [0,1,0,2,1,0,1,3,2,1,2,1]return 6 



```
# Your code goes here
def flooded_island(nums):
  n = len(nums)
  left = [0]*(n+1)
  for i in range(n):
    left[i+1] = max(left[i], nums[i])

  right = [0]*(n+1)
  for i in range(n):
    right[n-i-1] = max(right[n-i], nums[n-i-1])

  print(left)
  print(right)

  sum = 0
  for i in range(n):
    sum += min(left[i+1], right[i]) - nums[i]

  return sum
```
